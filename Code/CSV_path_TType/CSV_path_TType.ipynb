```python
import pandas as pd

# Cargar los archivos CSV
df1 = pd.read_csv("C:\\Users\\maseo\\OneDrive\\Escritorio\\GalaxyClassification\\Data\\Dataset\\NairAbrahamMorphology.csv", sep=';')
df2 = pd.read_csv("C:\\Users\\maseo\\OneDrive\\Escritorio\\GalaxyClassification\\Data\\Dataset\\fcP_ID.csv", sep=';')

# Fusionar los DataFrames en función de la columna "ID"
nuevo_df = pd.merge(df1, df2, on="spID")

# Me quedo con las columnas que me interesan
nuevo_df = nuevo_df[['#JID', 'fpCid', 'TType']]
# Guardar el nuevo DataFrame en un archivo CSV
nuevo_df.to_csv("C:\\Users\\maseo\\OneDrive\\Escritorio\\GalaxyClassification\\Data\\Dataset\\path_TType.csv", index=False, sep=';')

def agregar_texto(celda):
    return f"..\..\Data\Images\{celda}.jpg"

# Aplicar la función a cada celda de la columna "Nombre"
nuevo_df["fpCid"] = nuevo_df["fpCid"].apply(agregar_texto)

# Guardar el DataFrame modificado en un nuevo archivo CSV
nuevo_df.to_csv("C:\\Users\\maseo\\OneDrive\\Escritorio\\GalaxyClassification\\Data\\Dataset\\path_TType.csv", index=False, sep=';')
```
